### Ь Carga del modelo entrenado
Se carga el modelo guardado desde el archivo generado en la notebook anterior (incluido en la entrega del tp). Tambien se carga el label encoder para interpretar las clases predichas.

```python
import joblib

model = joblib.load('f1_predictor_model.joblib')
print('Modelo cargado correctamente.')
le = joblib.load("label_encoder.joblib")
print('Label encoder cargado correctamente.')
```

##  Predicci贸n sobre un ejemplo manual (ingresado por el usuario)

Aqu铆 se definen los valores de las caracter铆sticas para la carrera/piloto cuyo resultado se desea predecir

```python
import pandas as pd

ejemplo = {
    'grid': 2,
    'elo_driver_before': 1700,
    'elo_constructor_before': 1750,
    'driver_points_before': 180,
    'driver_wins_before': 5,
    'constructor_points_before': 350,
    'constructor_wins_before': 8,
    'qual_position': 2,
    'qualifying_gap': -0.120,
    'qual_gap_vs_teammate': -0.240,
    'race_gap_vs_teammate': -0.500,
    'year': 2024,
    'circuitId': 18,
    'AirTemp': 28.0,
    'TrackTemp': 42.0,
    'Humidity': 55.0,
    'Rainfall': 0.0,
    'WindSpeed': 4.0
}

features = [
    'grid',
    'elo_driver_before','elo_constructor_before',
    'driver_points_before','driver_wins_before',
    'constructor_points_before','constructor_wins_before',
    'qual_position','qualifying_gap','qual_gap_vs_teammate',
    'race_gap_vs_teammate',
    'year','circuitId',
    'AirTemp','TrackTemp','Humidity','Rainfall','WindSpeed'
]

ejemplo_df = pd.DataFrame([ejemplo])[features]

pred = model.predict(ejemplo_df)[0]
clase = le.inverse_transform([pred])[0]

print("Predicci贸n del modelo:", clase)
```
